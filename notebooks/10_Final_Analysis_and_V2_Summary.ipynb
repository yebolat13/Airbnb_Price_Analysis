```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

# Set a consistent style for plots
sns.set_style('whitegrid')
```

```python
# Create a dictionary to store all model results
results_data = {
    'Model': ['Linear Regression (V1)', 'Linear Regression (V1)', 'Linear Regression (V1)',
              'Random Forest (V2)', 'Random Forest (V2)', 'Random Forest (V2)'],
    'City': ['Berlin', 'Istanbul', 'Munich',
             'Berlin', 'Istanbul', 'Munich'],
    'RMSE': [93.31, 211.71, 142.19,
             48.90, 177.69, 103.06],
    'R2': [0.008, 0.022, -0.002,
           0.73, 0.31, 0.47]
}

# Create a DataFrame from the results data
results_df = pd.DataFrame(results_data)

print("--- All Model Results ---")
print(results_df)
```

```python
# Visualize RMSE comparison
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# RMSE Bar Plot
sns.barplot(data=results_df, x='City', y='RMSE', hue='Model', ax=ax[0])
ax[0].set_title('Model RMSE Comparison', fontsize=16)
ax[0].set_ylabel('RMSE', fontsize=12)
ax[0].set_xlabel('City', fontsize=12)
ax[0].legend(title='Model', loc='upper right')

# R-squared Bar Plot
sns.barplot(data=results_df, x='City', y='R2', hue='Model', ax=ax[1])
ax[1].set_title('Model R-squared (R²) Comparison', fontsize=16)
ax[1].set_ylabel('R²', fontsize=12)
ax[1].set_xlabel('City', fontsize=12)
ax[1].legend(title='Model', loc='upper right')

plt.tight_layout()
plt.show()
```

# Final Analysis and Conclusion (V1 vs. V2)

The project's journey from a simple baseline model to a more advanced solution highlights a critical aspect of data science: the importance of model selection. Our final analysis and the accompanying visualizations clearly demonstrate the dramatic performance improvement achieved in the "V2" phase.

## V1: Linear Regression Baseline Model

As shown in the initial analysis, the Linear Regression model performed poorly across all cities. The R² scores were near zero, and even negative for Munich, indicating that the model failed to capture the underlying patterns in the data. The high RMSE values suggested that the model's predictions were inaccurate.

## V2: Random Forest Advanced Model

By transitioning to a more powerful **Random Forest Regressor**, we observed a significant leap in predictive capability:

-   **Berlin:** The R² score jumped from a negligible 0.008 to a strong **0.73**, while the RMSE was nearly halved from 93.31 to **48.90**.
-   **Istanbul:** The R² score improved from 0.022 to **0.31**, and the RMSE dropped from 211.71 to **177.69**.
-   **Munich:** The most notable improvement was in Munich, where the R² score went from a negative value to a respectable **0.47**, and the RMSE was reduced from 142.19 to **103.06**.

## Overall Conclusion

The project successfully demonstrates a full data science workflow, starting with a simple baseline to validate the data pipeline, and then iteratively improving the solution. The Random Forest model proved to be far more suitable for this problem, as it is better equipped to handle the non-linear relationships and high dimensionality of the Airbnb dataset. The final results provide a strong, data-driven solution for price prediction and a clear roadmap for further enhancements.
