```python
# Import necessary libraries for data manipulation
import pandas as pd
import os

# ---
# 1. Loading the Datasets
# ---

# Define the project's root directory to ensure the code is portable
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
data_path = os.path.join(project_root, 'data')

# Load the Berlin dataset
berlin_file_path = os.path.join(data_path, 'berlin', 'raw', 'berlin_listings.csv.gz')
df_berlin = pd.read_csv(berlin_file_path, compression='gzip')

# Load the Istanbul dataset
istanbul_file_path = os.path.join(data_path, 'istanbul', 'raw', 'istanbul_listings.csv.gz')
df_istanbul = pd.read_csv(istanbul_file_path, compression='gzip')

# Load the Munich dataset
munich_file_path = os.path.join(data_path, 'munich', 'raw', 'munich_listings.csv.gz')
df_munich = pd.read_csv(munich_file_path, compression='gzip')

# ---
# 2. Initial Data Inspection
# ---

# Display the first 5 rows of each DataFrame to confirm successful loading
print("Berlin Dataset (Head):")
print(df_berlin.head())

print("\nIstanbul Dataset (Head):")
print(df_istanbul.head())

print("\nMunich Dataset (Head):")
print(df_munich.head())

# Get summary information about the datasets (columns, data types, non-null counts)
print("\nBerlin Info:")
df_berlin.info()

print("\nIstanbul Info:")
df_istanbul.info()

print("\nMunich Info:")
df_munich.info()
```
